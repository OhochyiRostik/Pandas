```python
# Від кожного елемента стовпчика віднімаєм min (map)
review_Ålder_min = reviews.Ålder.min()
reviews.Ålder.map(lambda p: p - review_Ålder_min)
```

```python
# Від кожного елемента стовпчика віднімаєм середнє по стовпчику (apply)
review_Ålder_mean = reviews.Ålder.mean()
def subtraction(row):
    row.Ålder = row.Ålder - review_Ålder_mean
    return row.Ålder

reviews.apply(subtraction, axis='columns')
```
