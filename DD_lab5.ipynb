```python
# Будуємо DataFrame з 6 та 7 стовпців, які використовуватимемо, як ознаки 
x=customer_set.iloc[:, [6,7]].values
```

### Етап побудови моделі

Реалізація алгоритму k-середніх для визначення оптимальної кількості кластерів. Завантажуємо Kmeans

```python
#імпортуємо бібліотеку і робило пустий список
from sklearn.cluster import KMeans
clalist=[]
```

Будемо використовувати метод Elbow для визначення k з параметром "k-means++"

```python
#inertia_ - це метод, який розділяє точки даних на кластери
# виконуємо кластеризацію за допомогою KMeans(n_clusters - кількість кластерів, 
#init - вибирає початкові центроїди кластера за допомогою вибірки на основі емпіричного розподілу ймовірностей внеску точок у загальну інерцію)
# random_state - Визначає генерацію випадкових чисел для ініціалізації центроїда. Використовуйте int, щоб зробити випадковість детермінованою
# fit - навчання
# inertia_ - Сума квадратів відстаней зразків до найближчого центру кластера, зважена за вагою зразків, якщо є.
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(x)
    clalist.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), clalist)
plt.title('Метод Elbow')
plt.xlabel('Кількість кластерів')
plt.ylabel('clalist')
plt.show()
```

```python
# fit_predict - навчання + передбачення
kmeansmodel = KMeans(n_clusters= 3, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(x)
```

### Етап візуалізації

```python
# cluster_centers_ - Координати центрів кластерів
plt.figure(figsize=(15,11.25))
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 40, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 40, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 40, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[0:6, 0], kmeans.cluster_centers_[0:6, 1], s = 200, c = 'yellow', label = 'Centroids')
plt.title('Кластери')
plt.xlabel('Social support')
plt.ylabel('Healthy life expectancy')
plt.legend()
plt.show()
```
