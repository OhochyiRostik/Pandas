```python
Rmean=movies.Rating.mean()
print("Середній рейтинг =",Rmean)
movies.loc[movies.Rating>Rmean]
```

```python
Rmean=movies.Rating.mean()
print("Середній рейтинг =",Rmean)
movies.loc[(movies.Rating<Rmean) & (movies.Year==2022)]
```

```python
movies.loc[(movies.Certificate=="PG-13") | (movies.Month=="December")]
```

```python
movies.loc[movies.Filming_location.isin(['USA'])]
```

```python
movies.loc[movies.Year.isin([2022,2020])]
```
